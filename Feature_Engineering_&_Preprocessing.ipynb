# Handling Class Imbalance

## What is Class Imbalance?

Class imbalance occurs when one class contains significantly more samples than the other class(es). In the IBM Employee Attrition dataset, approximately **83.9%** of employees did **not** leave the company, while only **16.1%** left. This creates an imbalanced binary classification problem.

---

## Why Can Class Imbalance Affect Machine Learning Models?

Machine learning models trained on imbalanced data tend to favor the majority class because it dominates the training data. As a result, the model may correctly predict most majority-class samples while failing to identify the minority class, which is often the class of greatest business interest.

---

## Why is Accuracy Alone Not Sufficient?

Accuracy can be misleading when classes are imbalanced. For example, if 84% of employees stay with the company, a model that predicts **"No Attrition"** for every employee would achieve approximately **84% accuracy**, despite failing to identify any employees who actually leave.

Therefore, accuracy alone does not provide a reliable measure of model performance for this problem.

---

## Which Evaluation Metrics are More Appropriate?

For imbalanced classification problems, the following evaluation metrics provide a more comprehensive assessment:

- **Precision:** Measures how many predicted attrition cases are actually correct.
- **Recall:** Measures how many actual attrition cases are correctly identified.
- **F1 Score:** Balances Precision and Recall into a single metric.
- **ROC-AUC Score:** Evaluates the model's ability to distinguish between employees who leave and those who stay across different classification thresholds.

These metrics provide a more reliable evaluation than accuracy alone.

---

## Should the Dataset be Balanced Before or After the Train-Test Split?

The dataset should be balanced **only after the train-test split**.

Balancing the dataset before splitting can introduce **data leakage**, allowing information from the test set to influence the training process. The correct workflow is:

1. Split the data into training and testing sets.
2. Apply balancing techniques (e.g., SMOTE or Random Oversampling) **only to the training data**.
3. Keep the test set unchanged so that it represents real-world data and provides an unbiased evaluation of model performance.
